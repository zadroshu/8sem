```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.distributions import Categorical
import numpy as np
import time
import random
from google.colab import files
```

```python
files.upload()
import kalah
```

```python
DL = 14
Prs = 14

model = 0
if os.path.exists('kallah_neyron.pt'):
  model = torch.load('kallah_neyron.pt')
else:
  model = torch.nn.Sequential(
          torch.nn.Linear(DL,Prs),
          torch.nn.Sigmoid(),
          torch.nn.Linear(Prs, 6),
          torch.nn.Softmax()
          )
```

```python
optimizer = torch.optim.Adam(model.parameters())
```

```python
def bolvan(x):
  rand_value = np.random.randint(6)
  while x[rand_value] == 0:
    rand_value = np.random.randint(6)
  return rand_value
```

```python
def space():
    return torch.tensor(kalah.game.board[0:])
```

```python
time_1 = time.time()

gamma = 1
alpha = 0.9
max_ep = 9999999999

for episode in range(max_ep):
  if episode % 1000 == 0:
    torch.save(model, 'kallah_neyron')
    
    tr = []
    kalah.game = kalah.Kalah()
    while not kalah.game.game_over():
      SPACE = space()

      if kalah.game.current_player == 0:
        probs = model(SPACE)
        m = Categorical(probs)
        action = m.sample()
        # print("Neyro first action hole value: {}:".format(SPACE[action]))
      else:
        action = bolvan(SPACE)
      # hole = int(input("Player {}'s turn. Enter hole number(0-5): ".format(kalah.game.current_player + 1)))
      # print("=========== Move ", kalah.game.move, " Current player {}: action {}: ===========".format(kalah.game.current_player, action))
      # print("player 2:", kalah.game.board[-1 * kalah.game.num_holes - 1], kalah.game.board[kalah.game.num_holes + 2:])
      # print("player 1:  ", kalah.game.board[:kalah.game.num_holes], kalah.game.board[kalah.game.num_holes])
      while True:
          # print("Player {}'s turn. Enter hole number(0-5): ".format(self.current_player+1))
          
          if kalah.game.board[action] != 0 and kalah.game.current_player == 0:
            reward = 1
            tr.append((SPACE.clone(), action.clone(), reward))
            break
          elif kalah.game.current_player == 0:
            reward = -100
            tr.append((SPACE.clone(), action.clone(), reward))
            # probs = model(SPACE)
            # m = Categorical(probs)
            # action = m.sample()
            kalah.game.board[kalah.game.num_holes + 1] = 72.
            kalah.game.get_winner()
            break
          else:
            break
      kalah.game.next_move(action)
    print("Game over. Player {} wins!".format(kalah.game.get_winner() + 1))
    
    loss = 0.
        
    T = len(tr)
    for t in range(T):
        R = 0.
        for i in range(t,T):
            R += (gamma**(i - t))*tr[i][2]

        loss += -alpha*R*Categorical(model(tr[t][0])).log_prob(tr[t][1])
        
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
        
        
print(time.time() - time_1)     
```
